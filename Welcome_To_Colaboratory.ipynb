```
import cv2
import numpy as np

# Read the Lenna image
img = cv2.imread('cameraman.tiff', cv2.IMREAD_GRAYSCALE)

# Apply uniform quantization
quantized_img = np.where(img > 128, 255, 0).astype(np.uint8)

# Save the resulting image
cv2.imwrite('quantized.tiff', quantized_img)

# Display the dimensions of the resulting image
print("Dimensions of the resulting image:", quantized_img.shape)
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the Lenna image
img = cv2.imread('cameraman.tiff', cv2.IMREAD_GRAYSCALE)

# Apply uniform quantization
quantized_img = np.where(img > 128, 255, 0).astype(np.uint8)

# Display the resulting image
plt.imshow(quantized_img, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```
